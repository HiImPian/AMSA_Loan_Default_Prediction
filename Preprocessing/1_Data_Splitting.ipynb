# AMSA: Data Splitting

## Relevant libraries

```python
import pandas as pd

from sklearn.model_selection import train_test_split # for splitting the data into train and test sets
```

## Data

```python
default_data = pd.read_csv('../data/Loan_default.csv')
default_data.head(n = 10)
```

checking for class imbalance

```python
default_unique_elements = [i for i in default_data["Default"].unique()] # storing all the unique values from the Default columns in our default_data's data frame
#print(default_unique_elements)

n = default_data["Default"].value_counts()
#print(n)

prop = n/len(default_data)
#print(prop)

target_class_distribution_df = pd.DataFrame(
    {
        "default": default_unique_elements, 
        "n": n, 
        "prop": prop
    }
)

target_class_distribution_df
```

Based on the table above, it is evident that we have a significant class imbalance issues that lie in the our data.

## Data splits

```python
default_X = default_data.drop(columns = ['Default'])
default_y = default_data['Default']
```

```python
X_train, X_test, y_train, y_test = train_test_split(default_X, default_y, test_size = 0.3, random_state = 1396, stratify = default_y) 
```

```python
print(f'X_train: {X_train.shape}')
print(f'X_test: {X_test.shape}')
print(f'y_train: {y_train.shape}')
print(f'y_test: {y_test.shape}')
```

```python
default_train = pd.concat([X_train, y_train], axis = 1)
default_test = pd.concat([X_test, y_test], axis = 1)
```

## Data exporting

```python
# saving the datasets onto the computer
default_train.to_csv(
    '../data/default_train.csv', 
    index = False, 
    sep = ','
)
```

```python
default_test.to_csv(
    '../data/default_test.csv',
    sep = ',',
    index = False
)
```
